```python
import matplotlib.pyplot as plt

# Dados
safra = ['202312', '202401', '202402']
total = [200, 300, 120]
com_saldo = [100, 120, 100]
faturas_nao_entregues = [10, 12, 2]
menor_que_15_reais = [2, 3, 1]

# Gráfico 1: Total de Safra ao longo do tempo
plt.figure(figsize=(10, 6))
plt.plot(safra, total, marker='o', linestyle='-', color='b')
plt.title('Total de Safra ao Longo do Tempo')
plt.xlabel('Safra')
plt.ylabel('Total')
plt.grid(True)
plt.show()
```

```python

# Gráfico 2: Comparação de Com Saldo e Faturas Não Entregues por Safra
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(safra))
plt.bar(index, com_saldo, bar_width, label='Com Saldo')
plt.bar(index, faturas_nao_entregues, bar_width, label='Faturas Não Entregues', bottom=com_saldo)
plt.xlabel('Safra')
plt.ylabel('Quantidade')
plt.title('Com Saldo vs Faturas Não Entregues por Safra')
plt.xticks(index, safra)
plt.legend()
plt.show()
```

```python

# Gráfico 3: Distribuição de Valores Menores que 15 Reais por Safra
plt.figure(figsize=(10, 6))
index = range(len(safra))
plt.bar(index, menor_que_15_reais, color='g')
plt.xlabel('Safra')
plt.ylabel('Quantidade')
plt.title('Valores Menores que 15 Reais por Safra')
plt.xticks(index, safra)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
safra = ['202312', '202401', '202402']
total = [200, 300, 120]
com_saldo = [100, 120, 100]
faturas_nao_entregues = [10, 12, 2]
menor_que_15_reais = [2, 3, 1]

# Gráfico 1: Total de Safra ao longo do tempo
plt.figure(figsize=(10, 6))
plt.plot(safra, total, marker='o', linestyle='-', color='b')
plt.title('Total de Safra ao Longo do Tempo')
plt.xlabel('Safra')
plt.ylabel('Total')
plt.grid(True)

# Adicionando etiquetas aos pontos de dados
for i, txt in enumerate(total):
    plt.annotate(txt, (safra[i], total[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()

```

```python
# Gráfico 2: Comparação de Com Saldo e Faturas Não Entregues por Safra
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(safra))
bar1 = plt.bar(index, com_saldo, bar_width, label='Com Saldo')
bar2 = plt.bar(index, faturas_nao_entregues, bar_width, label='Faturas Não Entregues', bottom=com_saldo)

plt.xlabel('Safra')
plt.ylabel('Quantidade')
plt.title('Com Saldo vs Faturas Não Entregues por Safra')
plt.xticks(index, safra)
plt.legend()

# Adicionando etiquetas aos dados
for bar in [bar1, bar2]:
    for rect in bar:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                     xy=(rect.get_x() + rect.get_width() / 2, height),
                     xytext=(0, 3),
                     textcoords="offset points",
                     ha='center', va='bottom')

plt.show()

```

```python
# Gráfico 3: Distribuição de Valores Menores que 15 Reais por Safra
plt.figure(figsize=(10, 6))
index = range(len(safra))
plt.bar(index, menor_que_15_reais, color='g')

plt.xlabel('Safra')
plt.ylabel('Quantidade')
plt.title('Valores Menores que 15 Reais por Safra')
plt.xticks(index, safra)

# Adicionando etiquetas aos dados
for i, txt in enumerate(menor_que_15_reais):
    plt.annotate(txt, (index[i], menor_que_15_reais[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```
